```python
import numpy as np
import pandas as pd

# Tính độ bao phủ (recall)
rec4 = np.sum([np.any(te[i, tpn3[i].astype(int)] != 0) for i in range(943)]) / 943
print(f"Độ bao phủ: {rec4}")
np.savetxt('recall.csv', [rec4], fmt='%f')  # Lưu recall vào file

# Tải danh sách phim
try:
    movies = pd.read_csv('ml-100k/u.item', sep='|', encoding='latin-1', header=None, usecols=[0, 1], names=['movie_id', 'movie_name'])
    movie_id_to_name = dict(zip(movies['movie_id'] - 1, movies['movie_name']))
    name_to_movie_id = {v: k for k, v in movie_id_to_name.items()}
except FileNotFoundError:
    print("Lỗi: Không tìm thấy file 'ml-100k/u.item'. Vui lòng tải MovieLens 100k và cập nhật đường dẫn.")
    exit()

# Hàm tìm top-N phim tương tự
def get_top_n_similar_movies(movie_name, nprobsim, name_to_movie_id, movie_id_to_name, n=10):
    if movie_name not in name_to_movie_id:
        return f"Phim '{movie_name}' không có trong dữ liệu. Vui lòng kiểm tra tên phim (ví dụ: 'Toy Story (1995)')."
    movie_idx = name_to_movie_id[movie_name]
    sim_scores = nprobsim[movie_idx]
    top_indices = np.argsort(sim_scores)[::-1][1:n+1]
    return [movie_id_to_name[idx] for idx in top_indices]

# Nhập tên phim từ terminal
movie_name = input("Nhập tên phim (ví dụ: Toy Story (1995)): ")
n = 5
top_n_movies = get_top_n_similar_movies(movie_name, nprobsim, name_to_movie_id, movie_id_to_name, n)
print(f"Top {n} phim tương tự '{movie_name}':")
for i, movie in enumerate(top_n_movies, 1):
    print(f"{i}. {movie}")

# 1. Star Wars (1977)
# 2. Aladdin (1992)
# 3. Lion King, The (1994)
# 4. Beauty and the Beast (1991)
# 5. Cinderella (1950)
```
